```python
Total_Value_By_Prd=df.groupby(['Brand'],as_index=False)['Value'].sum().sort_values(by=['Value'],ascending=False)
```

```python
Total_Value_By_Prd
```

```python
sale_gen=df.groupby(['Brand'],as_index=False)['Value'].sum().sort_values(by=['Value'],ascending=False)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=sale_gen,x='Brand',y='Value')
```

```python
Total_Value_By_Hospital=df.groupby('Hospital',as_index=False)['Value'].sum().sort_values(by='Value',ascending=False)
Total_Value_By_Hospital
```

```python
ILF_AND_LSF_DATA=df[df['Category']=='ILF&LSF']
```

```python
ILF_AND_LSF_DATA
```

```python
TOTAL_LSF_AND_ISF_VAL=ILF_AND_LSF_DATA.groupby('Brand',as_index=False)['Value'].sum().sort_values(by='Value',ascending=False)
TOTAL_LSF_AND_ISF_VAL
```

```python
TOTAL_LSF_AND_ISF_VAL=ILF_AND_LSF_DATA.groupby('Brand',as_index=False)['Value'].sum().sort_values(by='Value',ascending=False)
plt.title('Lsf and Ilf Value By Brand')
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=TOTAL_LSF_AND_ISF_VAL,x='Brand',y='Value')
```

```python
Total_Qty_Of_Lsf_And_Ilf_by_Brand=ILF_AND_LSF_DATA.groupby('Brand',as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
Total_Qty_Of_Lsf_And_Ilf_by_Brand
```

```python
Total_Qty_Of_Lsf_And_Ilf_by_Prd_and_Brand=ILF_AND_LSF_DATA.groupby(['Products','Brand'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
Total_Qty_Of_Lsf_And_Ilf_by_Prd_and_Brand
```

```python
Total_Qty_Of_Lsf_And_Ilf_by_Brand=ILF_AND_LSF_DATA.groupby('Brand',as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
plt.title('Lsf and Ilf Qty By Brand')
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=Total_Qty_Of_Lsf_And_Ilf_by_Brand,x='Brand',y='Qty')
```

```python
Total_Qty_Of_Lsf_And_Ilf_by_Prd_and_Brand=ILF_AND_LSF_DATA.groupby(['Products','Brand'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

sns.set(rc={'figure.figsize':(12,5)})
sns.barplot(data = Total_Qty_Of_Lsf_And_Ilf_by_Prd_and_Brand, x = 'Products',y= 'Qty', hue='Brand')
```

```python

plt.figure(figsize=(14, 6))
plt.pie(Total_Qty_Of_Lsf_And_Ilf_by_Brand['Qty'],labels=Total_Qty_Of_Lsf_And_Ilf_by_Brand['Brand'],autopct='%.2f%%',textprops={'fontsize': 10})

plt.title('Brand By Quantity')
plt.show()
```
